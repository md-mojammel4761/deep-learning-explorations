```python
path = 'image/download.jpg'
BGR_RGB = cv2.imread(path, cv2.IMREAD_COLOR)
b, g, r = cv2.split(BGR_RGB)
plt.figure(figsize=(16,60),dpi=150)

plt.subplot(141);plt.imshow(r);plt.title('Red colors')
plt.subplot(142);plt.imshow(g);plt.title('Grenn Color')
plt.subplot(143);plt.imshow(b);plt.title('Blue Color')

merged_img = cv2.merge((b, g, r))
plt.subplot(144);plt.imshow(merged_img[: ,: , ::-1]);plt.title('Merged Image')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
print(f'Row X Columns: {df.shape}')
```

```python
df.info()
```

```python
df.describe()
```

```python
value = df['survived'].value_counts()

plt.figure(figsize=(12,8))
plt.pie(
    value,
    labels=value.index,
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.05] * len(value),
    shadow=True,
    colors=plt.cm.Set2.colors
)
plt.title('Survived Distribution',fontsize=16,weight='bold')
plt.legend(title="Survived", labels=["No (0)", "Yes (1)"])
```

```python
sns.set_theme(style='whitegrid')

plt.figure(figsize=(8,5))
sns.countplot(
    x='sex',
    hue='survived',
    data=df,
    palette='Set1'
)
plt.title('Sex Distribution by Survival Status(Titanic)',fontsize=16,weight='bold')
plt.xlabel('Sex (Gender)',fontsize=16,weight='bold')
plt.ylabel('Count',fontsize=16,weight='bold')
plt.xticks(rotation=0)
plt.legend(title='Survived', labels=['No  (0)', 'Yes (1)'])
plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,5))
sns.countplot(
    x='pclass',
    hue='survived',
    data=df,
    palette='Set2'
)
plt.title('Pclass Distribution by Survival Status(Titanic)',fontsize=16,weight='bold')
plt.xlabel('Pclass Passenger (1 → First Class, 2 → Second Class, 3 → Third Class)',fontsize=12,weight='bold')
plt.ylabel('Number of Passenger',fontsize=12,weight='bold')
plt.xticks(rotation=0)
plt.legend(title='Survived',labels=['No  (0)','Yes (1)'])
plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,5))
sns.countplot(
    x='embarked',
    hue='survived',
    data=df,
    palette='Set3'
)
plt.title('Embarked Distribution by Survival status(Titanic)',fontsize=16,weight='bold')
plt.xlabel("Embarked Port (C = Cherbourg, Q = Queenstown, S = Southampton)", fontsize=12)
plt.ylabel("Passenger Count", fontsize=12)
plt.legend(title="Survived", labels=["No (0)", "Yes (1)"])

plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['age'],bins=30,kde=True)
```

```python
sns.histplot(data=df,x='age',hue='survived',bins=30,kde=True)
```

```python
sns.set_theme(style="whitegrid", palette="pastel")
sns.boxplot(
    x='survived',
    y='fare',
    data=df,
    palette='Set2',
    width=0.6
)
plt.title('Fare Distribution by Survival Status(Titanic)',fontsize=20,fontdict={'fontweight': 'bold'})
plt.xlabel('Survival Status (0 = Not Survived, 1 = Survived)',fontsize = 12,weight='bold')
plt.ylabel('Fare Paid',fontsize=12,weight='bold')
```

```python
sns.set_theme(style='white')

plt.figure(figsize=(10,8))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm',
    linewidths=0.5,
    linecolor='white',
    cbar_kws={'shrink':0.8}
)
plt.title('Correlation Heatmap (Titanic Dataset)',fontsize=16,weight='bold')
plt.xticks(rotation=45,ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
